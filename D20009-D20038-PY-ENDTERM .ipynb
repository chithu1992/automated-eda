<b>Q1)You will create a function named ‘Histogram’ which will take the entire dataset as input and  return the histograms for all the numerical variables in your dataset as .png files in your working  directory. Make sure that all the generated graphs have proper titles and axis labels
Histogram(data)  
-> returns histograms for all numerical variables in data. </b>

```python
#This function takes the dataframe and plots the histogram for numerical variables(Both continuous and discrete)
def Histogram_1(df):
    #Defining all types of numeric variables so that code is scalable to any data
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    #Taking copy so that original dataframe doesnt get hampered
    current_df_1 = df.copy(deep=True)
    #Creating a dataframe with only numeric variables
    df_numeric_variables = current_df_1.select_dtypes(include=numerics)
    #Plotting a histogram for each numeric column and saving the figure in the working directory
    for idx, variable in enumerate(df_numeric_variables.columns):
        df_numeric_variables.hist(column = variable,grid = False,figsize = (6,4),bins = 7,color ="turquoise")
        plt.xlabel(variable,fontsize = 12)
        plt.ylabel("Frequency",fontsize = 12)
        plt.title("Histogram of {}".format(variable),fontsize = 16)
        plt.savefig('Histogram of_{}.png'.format(variable),dpi=300,bbox_inches='tight')
        plt.show()
```

```python
Histogram_1(df)
```

```python
#missing value detection and correction
def missing_values(df):
#checking missing values
    missing = input("Do you want to check if missing values are present?Say y/n")
    if missing == "y":                
        missing_values = df.isnull().sum()
        print("**************************************************")
        print("----------------------------------------")
        print("No: of missing values for the variables")
        print("----------------------------------------")
        print(missing_values)
        
#correcting missing values        
    if missing == "n" or missing == "y":  
        missing_correction = input("Do you want to do missing value treatment?Say y/n")
        if missing_correction == "y":
        #filling categorical missing values if any
            df = df.fillna(df.mode().iloc[0])
            #filling numerical missing values if any
            df = df.fillna(df.mean())
            print("----------------------------------------")
            print("Checking whether missing values are present after missing value treatment for both non-numeric and numeric columns")
            missing_values_treated_check = df.isnull().sum()
            print("no: of missing values : ")
            print(missing_values_treated_check)
            print("----------------------------------------")
        if missing_correction == "n" or missing_correction == "y":
            print("**************************************************")
            df_duplicate_correction = duplicated_rows(df)
            return df_duplicate_correction
            





```
